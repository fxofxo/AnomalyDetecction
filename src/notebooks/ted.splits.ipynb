```python
import scipy.io
import numpy as np

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

## Shows windowing splits

```python
with open('anomaly-detection/splits.csv','rb') as fin:
    data = [ np.asarray(line.strip().split(b',')[1:], float) for idx,line in enumerate(fin)] # if idx != 0:
```

```python
n_rows=20
n_cols=10
plt.close('all')
fig, ax = plt.subplots(n_rows,n_cols,sharex='col', sharey='row',figsize=(20,20)) #,figsize=(3,3)
x=range(0,len(data[0]))

for n in range(0,200):
  
   ax[n // n_cols ,n%n_cols ].plot(x,data[n*1])
    
plt.show(fig)
```
